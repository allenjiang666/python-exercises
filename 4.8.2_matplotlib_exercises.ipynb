$$
y = x^2 - x -2 
$$

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
x = list(range(-10, 10))
y = [x ** 2 - x - 2 for x in x]
plt.plot(x, y)
plt.text(0, 2, '(0, 0)', fontsize=10, color='blue')
```

3.
Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize=(16, 9))

plt.subplot(2, 2, 1)
x = list(range(0,100))
y = [math.sqrt(x) for x in x]
plt.plot(x, y)
plt.title("$y=√x$")

plt.subplot(2, 2, 2)
x = list(range(0,10))
y = [x ** 3 for x in x]
plt.plot(x, y)
plt.title('$y = x^3$')

plt.subplot(2, 2, 3)
x = range(0, 10)
y = [2 ** x for x in x]
plt.plot(x, y)
plt.title('$y = x^2$')

plt.subplot(2, 2, 4)
x = range(-500, 500)
y = [math.tan(x) for x in x]
plt.plot(x, y)
plt.title('$y = tan(x)$')
```

```python
plt.figure(figsize=(16, 9))

x1 = list(range(1,100))
y1 = [math.sqrt(x1) for x1 in x1]

x2 = list(range(0,100))
y2 = [x2 ** 3 for x2 in x2]

x3 = range(0,100 )
y3 = [2 ** x3 for x3 in x3]

x4 = range(0, 100)
y4 = [math.tan(x4) for x4 in x4]

plt.plot(x1, y1, label="$y=√x$")
plt.plot(x2, y2, label='$y = x^3$')
plt.plot(x3, y3, label='$y = 2^x$')
plt.plot(x4, y4, label='$y = tan(x)$')
plt.legend()
```

```python
import numpy as np 

# .linspace arguments are (start, end, number_of_steps)
x = np.linspace(-2 * math.pi, 2 * math.pi, 1000)

# Isn't this neat?! np.tan(x) is similar to [math.tan(n) for n in x]
y = np.tan(x)

# This operation inserts a NaN where the difference between successive points is negative
# NaN means "Not a Number" and NaNs are not plotted or connected
# I found this by doing a search for "How to plot tan(x) in matplotlib without the connecting lines between asymtotes"
y[:-1][np.diff(y) < 0] = np.nan

# show grid
plt.grid()

plt.xlabel("x")
plt.ylabel("$tan(x)$")

# Set the x and y axis cutoffs
plt.ylim(-10,10)
plt.xlim(-2 * math.pi, 2 * math.pi)

# x_labels in radians
# For a more programmatic approach to radians, see https://matplotlib.org/3.1.1/gallery/units/radian_demo.html
radian_multiples = [-2, -3/2, -1, -1/2, 0, 1/2, 1, 3/2, 2]
radians = [n * math.pi for n in radian_multiples]
radian_labels = ['$-2\pi$', '$-3\pi/2$', '$\pi$', '$-\pi/2$', '0', '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$']

plt.xticks(radians, radian_labels)

plt.title("$y = tan(x)$", fontsize=14)
plt.plot(x, y)
plt.show()
```
