## Curves to graph
$y = 0x + 1$ and label the curve "O(1)"
$y = log(x)$ and label the curve "O(log n)"
$y = x$ and label the curve "O(n)"
$y = x * log(x)$ and label it "O(n log n)"
$y = x^2$ and label it "O(n^2)"
$y = 2^x$ and label it "O(2^n)"
$y = n!$ and label is "O(n!)"
$y = n^n$ and label it "O(n^n)"

```python
import matplotlib.pyplot as plt
import math
```

```python
plt.figure(figsize = (16, 9))

plt.title("Big O Notation", fontsize = 25, fontweight='bold')
plt.xlabel("Elements", fontsize = 20)
plt.ylabel("Operations", fontsize = 20)


x1 = [x/10 for x in range(30)]
y1 = [0 * x1 + 1 for x1 in x1]
plt.plot(x1, y1, label = "O(1)")

x2 = [x/10 for x in range(10, 30)]
y2 = [math.log(x2) for x2 in x2]
plt.plot(x2, y2, label = "O(log n)")

x3 = [x/10 for x in range(30)]
y3 = [x3 for x3 in x3]
plt.plot(x3, y3, label = "O(n)")

x4 = [x/10 for x in range(10,30)]
y4 = [x4 * math.log(x4) for x4 in x4]
plt.plot(x4, y4, label = "O(n log n)")

x5 = [x/10 for x in range(30)]
y5 = [x5 ** 2 for x5 in x5]
plt.plot(x5, y5, label =" O(n^2)")

x6 = [x/10 for x in range(30)]
y6 = [2 ** x6 for x6 in x6]
plt.plot(x6, y6, label = "O(x^n)")

x7 = range(3)
y7 = [math.factorial(x7) for x7 in x7]
plt.plot(x7, y7, label = "O(n!)")



x8 = [x/10 for x in range(30)]
y8 = [x8 ** x8 for x8 in x8]
plt.plot(x8, y8, label = "O(n^n)")

plt.legend(fontsize = 15)
```

```python
plt.figure(figsize = (16, 9))
x2 = [x/10 for x in range(10, 10**5)]
y2 = [math.log(x2) for x2 in x2]
plt.plot(x2, y2, label = "O(log n)")
plt.yscale("log")
```
